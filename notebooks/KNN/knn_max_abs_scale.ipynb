```python
X = df.iloc[:, 1:]
y = df.iloc[:,:1]
```

# Resampling

```python
from imblearn.over_sampling import SMOTE

resample = SMOTE(sampling_strategy='minority')
X, y = resample.fit_sample(X, y)
```

# Split The Dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0, stratify= y)
```

# Feature Scaling

# Max Abosulate Scaling

```python
from sklearn.preprocessing import MaxAbsScaler

scaler = MaxAbsScaler().fit(X_train)

X_train = scaler.transform(X_train)
X_train = pd.DataFrame(X_train)

X_test = scaler.transform(X_test)
X_test = pd.DataFrame(X_test)


# Setting Column Names from dataset
X_train.columns = X.columns
X_test.columns = X.columns
```

# KNN - K Nearest Neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
y_test_pred = knn.predict(X_test)
tp, fn, fp, tn = confusion_matrix(y_test, y_test_pred).ravel()
precision = (tp)/ (tp + fp)
recall =  tp/ (tp + fn)
print("True Negative:", tn)
print("True Positve:", tp)
print("False Positive:", fp)
print("False Negative:", fn)

print("Precision: ", (tp)/ (tp + fp))
print("Recall: ", tp/ (tp + fn))
print("Accuracy: ", (tp + tn)/(tp + fp + tn + fn))
print("F1 Score:" ,  2 * (precision * recall) / (precision + recall))
knn_f1 = 2 * (precision * recall) / (precision + recall)
```

```python
from sklearn.metrics import classification_report
y_test_pred = knn.predict(X_test)
print(classification_report(y_test, y_test_pred))
```

```python
from yellowbrick.classifier import confusion_matrix

plt.figure(figsize=(6, 5))
visualizer = confusion_matrix(
    knn,
    X_test, y_test,
    is_fitted=True
    
)
visualizer.show();
```

## Cross validation and Learning Curve

```python
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold(
    n_splits=10,random_state=0,
    shuffle=False
)
```

```python
from yellowbrick.model_selection import LearningCurve

plt.figure(figsize=(15, 10))
visualizer = LearningCurve(
    model= knn, 
    cv=cv, 
    scoring='f1_weighted', 
    train_sizes=np.linspace(0.3, 1.0, 10), 
    n_jobs=4,
    is_fitted=True,
    stratify = True
)
visualizer.fit(X, y);
visualizer.show();
```

```python
from yellowbrick.model_selection import CVScores
visualizer = CVScores(knn, cv=cv, scoring='f1_weighted', is_fitted=True, stratify = True)

visualizer.fit(X, y)        # Fit the data to the visualizer
visualizer.show()           # Finalize and render the figure
```

```python
from yellowbrick.classifier import ROCAUC

plt.figure(figsize=(15, 10))
visualizer = ROCAUC(
    knn,
    is_fitted=True,
)
visualizer.score(X_test, y_test)
visualizer.show();
```
